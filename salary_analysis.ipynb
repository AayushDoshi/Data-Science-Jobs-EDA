1 Which role has the highest salary employment wise?

```python
df_q1=df[['employment_type','job_title','salary_in_usd']]
q1 = df_q1.groupby(['employment_type', 'job_title']).agg(avg_sal=('salary_in_usd', 'mean')).rename(columns={'avg_sal': 'mean_salary'}).sort_values(by='mean_salary',ascending=True).reset_index()
q1_top1=q1.loc[q1.groupby('employment_type')['mean_salary'].idxmax()].sort_values(by='mean_salary', ascending=False)
q1_top1
```

4 Which countries pay the highest for which roles?

```python
q4=df.groupby(['company_location','job_title']).agg(mean=('salary_in_usd','mean')).rename(columns={'mean':'salary_mean'}).sort_values('salary_mean',ascending=False).reset_index()
q4_top1 = q4.loc[q4.groupby('company_location')['salary_mean'].idxmax()]
q4_top1
```

6 Which experience level has the highest hiring?

```python
q6=df.groupby('experience_level').agg({'experience_level':'count'}).rename(columns={'experience_level':'no_of_hired'}).sort_values('no_of_hired',ascending=False).reset_index()
q6
```

7 Does company size affect the rate of hiring and pay scale?

```python
q7_hired=df.groupby(['company_size','work_year']).agg({'work_year':'count'}).rename(columns={'work_year':'no_of_hired'}).sort_values('company_size')
q7_hired['hiring_rate']=round(q7_hired.groupby('company_size')['no_of_hired'].pct_change()*100, 2)
q7_hired
```

```python
q7_sal=df.groupby(['company_size','work_year']).agg({'salary_in_usd':'mean'}).rename(columns={'salary_in_usd':'mean_sal'}).sort_values('company_size')
q7_sal['growth_rate']=round(q7_sal.groupby('company_size')['mean_sal'].pct_change()*100, 2)
q7_sal
```

8 What is the year over year (YoY) salary growth at different levels?

```python
q8 = df.groupby(['experience_level', 'work_year']).agg(mean_sal=('salary_in_usd', 'mean')).round()
q8['salary_growth'] = round(q8.groupby('experience_level')['mean_sal'].pct_change() * 100, 2)
q8
```
